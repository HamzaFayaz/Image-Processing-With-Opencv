```python
''' 






'''

'''
Using the color Images 
'''

hsv_img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2.imshow("orignal1",hsv_img)

h , s, v = cv2.split(hsv_img)

v=cv2.equalizeHist(v) 




merged_hsv = cv2.merge((h,s,v))

bgr_img = cv2.cvtColor(merged_hsv,cv2.COLOR_HSV2BGR)

cv2.imshow("orignal",img)
cv2.imshow("output",bgr_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Applying The Clahe Method for image enhancemant

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)

```

```python
# Steps :  Make a clahe  and then apply of the image  like mask making and apply


clahe = cv2.createCLAHE(clipLimit=5) # Enter the pixel values threshold set for chale for details check the chale details above

clahe_img = clahe.apply(gray_img)
cv2.imshow("orignal",gray_img)
cv2.imshow("output",clahe_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
