```python
import cv2
```

```python
img = cv2.imread("fig1.jpg")
img = cv2.resize(img,(400,400))

img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# sobal_x = cv2.Sobel(img,ddepth=-1,dx=1,dy=0)
# sobal_y = cv2.Sobel(img,ddepth=-1,dx=0,dy=1)
sobal_x = cv2.Sobel(img,-1,1,0)
sobal_y = cv2.Sobel(img,-1,0,1)

sobal_xy = cv2.addWeighted(sobal_x,0.5,sobal_y,0.5,0) # 0.5 , 0.5 and 0 is the values for (visiblity)

Lap = cv2.Laplacian(img , -1)

# canny = cv2.Canny(img , threshold1=80 , threshold2= 150 )  the threshold is for 4 number step 

canny = cv2.Canny(img , 80 , 150 )


```

```python
cv2.imshow("Sobal X ",sobal_x)
cv2.imshow("Sobal y ",sobal_y)
cv2.imshow("Sobal X+Y",sobal_xy)
cv2.imshow("Laplacian ",Lap)
cv2.imshow("canny ",canny)
cv2.imshow("Img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
